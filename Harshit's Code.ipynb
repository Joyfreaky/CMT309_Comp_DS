## Q2 Pluralizing a given word

The aim is to check the given word and pluralize.
If the word is in the given proper noun form then the output should not be changed, for this we check in already proper noun file which is given.
If the word is already in plural then return the word without change, this is done by checking whether the word is ening with s.
If the word ends with vowels then add s as suffix, by checking the last digit using regex whether it is present in vowels.
If the word ends with y and previous to y is in vowels then add ies as suffix.
If the word ends with sh|ch|z the change it with es as suffix, this is done by giving a or condition to check the suffix.
If the word ends with f then the changed word sould be ves as suffix.

```python
import re
def pluralize(word):
    
  word.lower()
  word_present_in_proper_noun = open("proper_nouns.txt" ,"r").readlines()    #reading the proper noun file
  
  output = ['Proper noun','Already in Plural','Empty String','Success']      #declaring the output status
  operation = output[3]
  word_present_in_proper_noun = [line.rstrip() for line in word_present_in_proper_noun]    #For the string in input stripping it into letters
 
  if word in word_present_in_proper_noun:                                    #Checking for proper noun
    operation=output[0]
    word_in_plural = word
  elif re.search('[s]$',word):                                               #checking whether the word ends with s
    word_in_plural = word
    operation=output[1]
  elif re.search('[aeiou]$',word):                                           #checking for vowels at the end of word
    word_in_plural = re.sub('$','s',word)
  elif re.search('[^aeiou]y$',word):                                         #checking for y at the end and any vowels before it
    word_in_plural = re.sub('y$','ies',word)
  elif re.search('[sh|ch|z]$',word):                                         #checking for word ending with sh or ch or z
    word_in_plural = re.sub('$','es',word)
  elif re.search('[f]$',word):                                               #checking for the word ending with f
    word_in_plural = re.sub('f$','ves',word)
  
  

  else:
    if word=='':                                                             #If any empty string is entered
      operation=output[2]
      word_in_plural =""
    else:
        word_in_plural = word +'s'
    operation = output[3]
        
  
  return{
      "plural":word_in_plural,
      "Status":operation
      }

### --- IMPORTANT: DO NOT REMOVE OR CHANGE THE CODE BELOW ---
TEST_CASES = """failure
food
Zulma
injury
elf
buzz
computers
PCs

highway
presentation
pouch
adam""".split('\n')

if __name__ == '__main__':

  for test_noun in TEST_CASES:

    print(test_noun,'-->',pluralize(test_noun))
    print('----')

### --- The outputs of test cases are given below. Please compare with your results.
# ************************************************************************************
# failure --> {'plural': 'failures', 'status': 'success'}
# ----
# food --> {'plural': 'foods', 'status': 'success'}
# ----
# Zulma --> {'plural': 'Zulma', 'status': 'proper_noun'}
# ----
# injury --> {'plural': 'injuries', 'status': 'success'}
# ----
# elf --> {'plural': 'elves', 'status': 'success'}
# ----
# buzz --> {'plural': 'buzzes', 'status': 'success'}
# ----
# computers --> {'plural': 'computers', 'status': 'already_in_plural'}
# ----
# PCs --> {'plural': 'PCs', 'status': 'already_in_plural'}
# ----
#  --> {'plural': '', 'status': 'empty_string'}
# ----
# highway --> {'plural': 'highways', 'status': 'success'}
# ----
# presentation --> {'plural': 'presentations', 'status': 'success'}
# ----
# pouch --> {'plural': 'pouches', 'status': 'success'}
# ----
# adam --> {'plural': 'adam', 'status': 'proper_noun'}
# ----
```
